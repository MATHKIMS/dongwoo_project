```python
# train, test set 분리

from sklearn.model_selection import train_test_split
x = ratings.copy()
y = ratings['user_id']

# 100000개 중 train data가 75000개, test data가 25000개
# stratify=y : x와 y를 train와 test로 나눌 때 같은 비율로 나누겠다는 의미
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y)
```
