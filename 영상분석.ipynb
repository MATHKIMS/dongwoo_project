```python
### packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
### 영상 출력 함수
def fn_imshow(img, axis='on', fig_size=''):
    if fig_size!='': plt.figure(figsize=fig_size)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    if axis!='on': plt.axis('off')
    plt.show()
```

```python
### 영상 생성
img = np.zeros((512,512,3), np.uint8) + 255 # ((),dtypes)
```

```python
### 영상 출력
fn_imshow(img, axis='on')
```

```python
### 두께가 5픽셀인 파란색 대각선
cv2.line (img,(0,0),(511,511),(255,0,0),5) # cv2.line (img,시작좌표,끝좌표,(B,G,R),5)

### 영상 출력
fn_imshow(img, axis='on')
```

'''■ 직사각형(rectangle)
직사각형의 왼쪽 위 모서리와 오른쪽 아래 모서리 좌표 사용
ㅇ rectangle() - OpenCV
cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType]])
img: 영상
pt1: 직사각형의 첫 번째 점(왼쪽 위 좌표)
pt2: 직사각형의 두 번째 점(오른쪽 아래 좌표)
color: 선 색상(BGR의 경우 튜플로 전달)
thickness: 선 두께(원과 같이 닫힌 도형에 -1 을 넘기면 도형을 채움. 기본 두께 = 1)
lineType: 라인의 유형(선종류. 기본값=8-connected. 곡선에 적합한 라인=cv2.LINE_AA)'''

```python
### 직사각형
cv2.rectangle(img,(384,0),(510,128),(0,255,0),3)

### 영상 출력
fn_imshow(img, axis='on')
```

```python
### 원
cv2.circle(img,(447,63), 63, (0,0,255), -1)

### 영상 출력
fn_imshow(img, axis='on')
```

■ 타원(ellipse)
중심 위치(x,y)와 축 길이(주축 길이, 부축 길이)
회전 각도와 타원 호의 시작과 끝
ㅇ ellipse() - OpenCV
cv.ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType]])
center: 중심 위치(x,y)
axes: 축 길이(주축 길이, 부축 길이)
angle: 회전 각도
startAngle, endAngle: 주축에서 시계 방향으로 측정된 타원 호의 시작과 끝

```python
### 타원
cv2.ellipse(img,(150,300),(100,50),45,0,180,255,2)

### 영상 출력
fn_imshow(img, axis='on')
```

```python
# ■ 글자(text)
# 글자와 글자를 배치하려는 위치 좌표(왼쪽 하단 모서리)
# 글꼴의 유형과 크기를 지정
# lineType = cv.LINE_AA 권장
# putText() - OpenCV
# cv.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType]])
# img: 영상
# text: 글자
# org: 영상에 있는 문자열의 왼쪽 아래 모서리 좌표
# fontFace: 글꼴 유형(Font type)
# fontScale: 글자 크기(글꼴별 기본 크기를 곱한 글꼴 배율 인수)
# color: 텍스트 색상
# thickness: 텍스트를 그리는 데 사용되는 선의 두께
# lineType: 라인 유형
```

```python
### 파일 - sudoku.png
img_file = 'sudoku.PNG'

### 영상 읽기
img_raw = cv2.imread(img_file)

### Grayscale
img_gray = cv2.cvtColor(img_raw, cv2.COLOR_BGR2GRAY)

### Canny edges
img_Canny = cv2.Canny(img_gray, 50, 150)
```

```python
### 출력 영상
img_out = img_raw.copy()

### 직선 검출
lines = cv2.HoughLines(img_Canny,        # 이진 영상
                       1,                # 거리 해상도(픽셀 단위)
                       np.pi/180,        # 각도 (라디안 단위)
                       200)              # 임계값
### rho, theta
#lines
```

```python
### 직선 표시
for i in range(len(lines)):
    for rho, theta in lines[i]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))
        cv2.line(img_out, pt1, pt2, (0,0,255), 2)
```

```python
### 영상 출력
titles = ["원래 영상", "Grayscale", "Canny edges", "직선 검출"]
images = [img_raw, img_gray, img_Canny, img_out]
for i in range(len(images)):
    img_rgb = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
    plt.subplot(2, 2, i+1), plt.imshow(img_rgb)
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
### 출력 영상
img_out = img_raw.copy()

### 직선 검출
lines = cv2.HoughLinesP(img_Canny, 1, np.pi/180, 50, None, 100, 10)

### 직선 표시
for i in range(len(lines)):
    for x1,y1,x2,y2 in lines[i]:
        cv2.line(img_out,(x1,y1),(x2,y2),(0,0,255),3)
```

```python
### 영상 출력
plt.imshow(cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
### 파일 - coins_2.jpg
img_file = 'coins_2.jpg'

### 영상 읽기
img_raw = cv2.imread(img_file)

### Grayscale
img_gray = cv2.cvtColor(img_raw, cv2.COLOR_BGR2GRAY)

### 가우시안 필터
img_Gauss = cv2.GaussianBlur(img_gray, (0, 0), 1)

### 원 검출
circles = cv2.HoughCircles(img_Gauss, cv2.HOUGH_GRADIENT, 1, 50,
                           param1=150, param2=40, minRadius=50, maxRadius=180)

### 원 좌표
circles = np.uint16(np.around(circles))
circles
```

```python
### 파일 - coins_2.jpg
img_file = 'coins.png'

### 영상 읽기
img_raw = cv2.imread(img_file)

### Grayscale
img_gray = cv2.cvtColor(img_raw, cv2.COLOR_BGR2GRAY)

### 가우시안 필터
img_Gauss = cv2.GaussianBlur(img_gray, (0, 0), 1)

### 원 검출
circles = cv2.HoughCircles(img_Gauss, cv2.HOUGH_GRADIENT, 1, 50,
                           param1=150, param2=40, minRadius=50, maxRadius=180)

### 원 좌표
circles = np.uint16(np.around(circles))
circles
```

```python
### 출력 영상
img_out = img_raw.copy()

### 원 표시
for cx, cy, radius in circles[0,:]:
    cv2.circle(img_out,(cx, cy), int(radius),(0,0,255),2)

### 영상 출력
plt.imshow(cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
### 출력 영상
img_out = img_raw.copy()

### 원 표시
for cx, cy, radius in circles[0,:]:
    cv2.circle(img_out,(cx, cy), int(radius),(0,0,255),2)

### 영상 출력
plt.imshow(cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
### 출력 영상
img_out = img_raw.copy()

### 원 표시 & 동전 합 계산
won_sum = 0
for cx, cy, radius in circles[0,:]:
    ### 반지름을 이용하여 동전을 구분
    if radius>120:
        won = 500
    elif radius>100:
        won = 100
    elif radius>80:
        won = 50
    elif  radius<=80:
        won = 10
    ### 동전 합
    won_sum += won
    ### 원 표시
    cv2.circle(img_out,(cx, cy), int(radius),(0,0,255),2)
    ### 동전 표시
    cv2.putText(img_out, str(won), (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (255, 0, 0), 2, cv2.LINE_AA)

### 동전 합 표시
cv2.putText(img_out, "Total: "+str(won_sum), (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 0), 4, cv2.LINE_AA)

### 영상 출력
plt.imshow(cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB))
plt.show()
```
